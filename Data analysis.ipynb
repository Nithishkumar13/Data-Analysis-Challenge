### SOME USEFUL FUNCTION

###### column access function
    The columns of the dataset can be accessed by the <b>display_columns</b> function. The columns names are appended in the list and by using this function. It returns the columns names in list. 

###### statistical values 
    Function called <b>addition</b> is used to add the element in the list or array by using for loop. function <b>mean</b> is used to add the element in the list or array by using for loop and which is divided by lenght of hte same list. function <b>std_dev</b> uses the formula of standard deviation. <b>percent</b> is used to return 25,50,75 percentile by using np.percentile().which takes two argument data and no.of percentile(For example:25,50,70).

##### indentify unique value:
    Function called <b>pick_unique</b> has need 1 parameter which may be a list or array. iteration was done with help of for loop.This fuction takes one value during iteration, if the item not in the uniq list it will be append. thus we got the final unique values.

```python
def display_columns(data):
    col=[]
    for i in data:
        col+=[i]
    return col
columns = display_columns(df)

def addition(a):
    add=0
    for i in np.array(a):
        add+=i
    return add

def mean(x):
    return addition(x)/len(x)

def std_dev(s):
    lst=[]
    for i in np.array(s):
        lst+=[pow(i-mean(s),2)]
    return np.sqrt(addition(lst)/len(s))

def percent(x,a):
        return np.percentile(x,a)
    
def pick_unique(data):
    uniq=[]
    for j in data.values:
        if j in uniq:
            pass
        else:
            uniq+=[j]
    return uniq
```

### get_info() function:
    this is a function written to collect the information about the columns, data types, no.of unique values.It takes the dataframe as an argument.This  function arrange the above mentioned data in a proper manner otherwise it is called a dataframe about data.

```python
def get_info(data):
    dtype,unique,col_sum,=[],[],[]
    col=display_columns(data)
    for i in data:
        count=0
        dtype+=[data[i].dtypes]
        
    ## for selecting unique element in the dataframe in each column
        for j in pick_unique(data[i]):
            count+=1
        unique+=[count]
        
    ## code for organizing the calculated data
    return pd.DataFrame(data={'column name':col,'data type':dtype,'unique value':unique},
                        index=range(1,len(col)+1))
get_info(df)
```

## get_statistical_data():
    It returns sum, mean, standard deviation, minimum, maximum, 25 percet, 50 percent, 75percent of the data which is passed as an argument. This is much useful to see the statistical informtion about the data. It is used in filling na_value by mean ,analyzing standard distribution of data, range of data, frequency of data etc... It give result to only numerical or float dtype column.

```python
def get_statistical_data(data):
    col,col_sum,col_mean,col_std,mini,maxi=[],[],[],[],[],[]
    per_25,per_50,per_75=[],[],[]
    col=display_columns(data)
    for i in data:
        if (data[i].dtypes=='int64') or (data[i].dtypes=='float64'):
            col_sum+=[addition(data[i])]
            col_mean+=[mean(data[i])]
            col_std+=[std_dev(data[i])]
            mini+=[min(data[i])]
            maxi+=[max(data[i])]
            per_25+=[percent(data[i],25)]
            per_50+=[percent(data[i],50)]
            per_75+=[percent(data[i],75)]
        else:
            col.remove(i)
    d={'column name':col,'sum':col_sum,'mean':col_mean,'std':col_std,
       'min':mini,'max':maxi,'25%':per_25,'50%':per_50,'75%':per_75}
    ## code for organizing the calculated data
    return pd.DataFrame(data=d,index=range(1,len(col)+1))
get_statistical_data(df)
```

## Check For Null:
    This is a function which is created to count the null values in the form of NaN in each columns of the dataset. it takes each and every value and check wheather it is equal to NaN or not if NaN add count. finally print the columns with null.
    function: check_for_null
    parameters: dataframe(d)

    
## Fill NaN:
    This function was writtern to fill the null values in the given column with the provided list with it
    function: fill_nan
    parameters: column with NaN(d) , list to be filled in NaN(fill)

```python
def replace(data,character,replace):
    for i in data:
        lst=[]
        for j in data[i]:
            if j == character:
                lst+=[replace]
            else:
                lst+=[j]
        data[i]=lst
    return data

def check_for_null(data):
    for i in data:
        count=0
        for j in data[i]:
            if j=='NaN':
                count+=1
        if count>0:
            print(i,"has {} null values".format(count))
    print("other columns has no null value")

def fill_nan(d,fill):
    lst=[]
    count=0
    for i in d:
        if i=='NaN':
            lst+=[fill[count]]
            count+=1
        else:
            lst+=[i]
    return np.array(lst)

```

# Justification of Our Analysis

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
## preparing variable

x = df.drop(['price','compression-ratio','symboling','num-of-doors'],axis=1)
y = df['price']
x= pd.get_dummies(x,drop_first=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
train_acc=[]

## training a model and select suitable accuracy 
for i in range(1,31):
    knn = KNeighborsRegressor(n_neighbors=i)
    model = knn.fit(x_train,y_train)
    pred = model.predict(x_train)
    train_acc += [r2_score(y_train,pred)]   

## visualization of accuracy
print('neighbor {} gives the good accuracy {}\n\n'.format(train_acc.index(max(train_acc))+1,max(train_acc)))
plt.plot(range(1,31),train_acc)
plt.scatter(range(1,31),train_acc)
plt.grid(True)
plt.xlabel('k nieghbor')
plt.ylabel('accuracy')
plt.title('variation in accuracy with respect to neighbor for training data')
plt.show()

## finding test accuracy
knn = KNeighborsRegressor(n_neighbors=train_acc.index(max(train_acc))+1)
model = knn.fit(x_train,y_train)
pred = model.predict(x_test)
print("test accuracy for same neighbor = ",r2_score(y_test,pred))
```
